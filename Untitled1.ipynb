```python
from keras.models import Sequential,Model
from keras.layers import Dense, Input, Conv2D, MaxPool2D,UpSampling2D,DepthwiseConv2D,BatchNormalization

from keras.models import Sequential,Model
from keras.layers import Dense, Input, Conv2D, MaxPool2D, Flatten
from keras.callbacks import EarlyStopping

from tensorflow.keras.layers import Conv2D,ReLU
```

#1 punto

Intput (10,)

Dense [1024,256,64,16] [relu,relu,relu,linear]

Output(1,)


```python
input_1=Input(shape=(10))

dense=Dense(1024,activation='relu')(input_1)
dense1=Dense(256,activation='relu')(dense)
dense2=Dense(64,activation='relu')(dense1)
dense3=Dense(16,activation='relu')(dense2)
dense3=Dense(1,activation='linear')(dense3)

model=Model(inputs=input_1,outputs=dense3)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model,show_shapes=True,show_layer_activations=True)
```
